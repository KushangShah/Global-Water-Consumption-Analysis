```python
# Function to plot top 10 countries for a given metric

df['Country'] = df['Country'].astype(str)

def plot_top_10(metric, title, xlabel):
    top_10 = df.groupby('Country')[metric].mean().nlargest(10).reset_index()

    plt.figure(figsize=(10, 6))
    sns.barplot(data=top_10, x=metric, y="Country", hue="Country", palette="viridis", legend=False)

    plt.xlabel(xlabel)
    plt.ylabel("Country")
    plt.title(title)
    plt.grid(axis="x", linestyle="--", alpha=0.7)
    plt.show()
```

```python
# char 1:- Top 10 Countries by Total Water Consumption
plot_top_10('Total Water Consumption (Billion Cubic Meters)', 
            'Top 10 Countries by Total Water Consumption', 
            'Total Water Consumption (Billion Cubic Meters)')
```

```python
# char 2:- Top 10 Countries by per Capita water use
plot_top_10("Per Capita Water Use (Liters per Day)",
            "Top 10 country by Per Capita Water Use",
            "Per Capita Water User(Liter per day)")
```

```python
# Chart 3:- Top 10 country V/S 'Agricultural Water Use (%)'
plot_top_10('Agricultural Water Use (%)',
            "Top 10 country by Agriculture water useage",
            "Agriculture Water Use")
```

```python
# Chart 4:- Top 10 country by 'Industrial Water Use (%)'
plot_top_10("Industrial Water Use (%)",
            "Top 10 Country by Industrial Water useage",
            "Industry Water Useage")
```

```python
# Chart 5:- Top 10 country by 'Household Water Use (%)'
plot_top_10("Household Water Use (%)",
            "Top 10 Country by HouseHold water useage",
            "HouseHold Water Useage")
```

```python
# Chart 6:- Top 10 country by low water scarcity
low_scarcity_df = df[df["Water Scarcity Level"] == "Low"]

top_10_low_scarcity = (
    low_scarcity_df.groupby("Country")["Total Water Consumption (Billion Cubic Meters)"]
    .mean()
    .nlargest(10)
    .reset_index()
)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_low_scarcity,
            x="Total Water Consumption (Billion Cubic Meters)",
            y="Country",
            hue="Country",
            palette="viridis",
            legend=False)

plt.xlabel("Total Water Consumption (Billion Cubic Meters)")
plt.ylabel("Country")
plt.title("Top 10 Countries with Low Water Scarcity")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Chart 7:- Top 10 country V/S High "Water Scarcity Level" 
high_scarcity = df[df['Water Scarcity Level'] == 'High']

top_10_high_scarcity = (
    high_scarcity.groupby("Country")["Total Water Consumption (Billion Cubic Meters)"]
    .mean()
    .nlargest(10)
    .reset_index()
)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_high_scarcity,
            x="Total Water Consumption (Billion Cubic Meters)",
            y='Country',
            hue='Country',
            palette='viridis',
            legend=False)
plt.xlabel("High Water Scarcity Level")
plt.ylabel("Top 10 country")
plt.title("Top 10 country by High Water scarcity level")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Chart 8:- Top 10 country V/S Medium "Water Scarcity Level" 
medium_scarcity = df[df['Water Scarcity Level'] == 'Medium']

top_10_medium_scarcity = (
    medium_scarcity.groupby("Country")['Total Water Consumption (Billion Cubic Meters)']
    .mean()
    .nlargest(10)
    .reset_index()
)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_medium_scarcity,
            x="Total Water Consumption (Billion Cubic Meters)",
            y="Country",
            hue='Country',
            palette='viridis',
            legend=False)
plt.xlabel("Medium Water scarcity level")
plt.ylabel("Top 10 country")
plt.title("Top 10 country by Medium Water scarcity level")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Chart 9:- Top 10 country V/S Rain fall impact
plot_top_10('Rainfall Impact (Annual Precipitation in mm)',
            "Top 10 country by Rainfall impact",
            "Average Annual Rain Fall impact in mm")
```

```python
# Chart 10:- water use distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[
    ['Agricultural Water Use (%)',
    'Industrial Water Use (%)',
    'Household Water Use (%)']
    ], palette='coolwarm')
plt.ylabel("Percentage")
plt.title("Distribution of Water Use by Sector")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

- Agriculture is the dominant water consumer, meaning policies for water conservation should focus here.
- Industrial and household water use are significantly lower, but they still need efficient management.

```python
# Chart 11:- relationship betweeen water consuption and groundwater depletion
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Total Water Consumption (Billion Cubic Meters)'],
                y=df['Groundwater Depletion Rate (%)'],
                hue=df['Water Scarcity Level'],
                palette='coolwarm',
                alpha=0.7)
plt.title("Total water consumption V/S GroundWater Depletion")
plt.xlabel("Total Water Consumption (Billion Cubic Meter)")
plt.ylabel("GroundWater Depletion Rate(%)")
plt.grid()
plt.show()
```

- The data points are widely scattered across all values of water consumption and depletion rate.
- This suggests that total water consumption alone may not be the sole factor driving groundwater depletion.

```python
# Chart 12:- Rainfall V/S water scarcity
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=['number'])
corr = numeric_df.corr()
sns.heatmap(data=corr,
            annot=True,
            cmap='coolwarm',
            fmt='.2f',
            linewidths=0.5)
plt.title("Correlation Heatmap of water usage matrics")
plt.show()
```

```python
# Chart 13:- Water consuption over time
plt.figure(figsize=(16, 6))

top_countries = df.groupby("Country")['Total Water Consumption (Billion Cubic Meters)'].mean().nlargest(5).index

for country in top_countries:
    country_data = df[df['Country'] == country].groupby('Year')['Total Water Consumption (Billion Cubic Meters)'].sum()
    plt.plot(country_data.index, country_data.values, label=country)

plt.title("Water Consumption Trend Over Time (Top 5 Countries)")
plt.xlabel("Year")
plt.ylabel("Total Water Consumption")
plt.legend()
plt.grid()
plt.show()
```

- The water consumption trend for each country varies significantly, with sharp ups and downs.
- This suggests that water consumption is not constant and is influenced by multiple factors like population growth, climate change, industrial demand, and government policies.

```python
# Chart 14:- Water scarcity V/S Population density
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Total Water Consumption (Billion Cubic Meters)'],
                y=df['Water Scarcity Level'],
                size=df['Per Capita Water Use (Liters per Day)'],
                hue=df['Water Scarcity Level'],
                sizes=(20, 400), alpha=0.7)
plt.title("water Scarcity V/S Total Water consumption")
plt.xlabel("Total Water Consumption")
plt.ylabel("Water Scarcity Level")
plt.grid()
plt.show()
```

```python
# chart 15:- Groundwater Depletion vs. Rainfall
plt.figure(figsize=(10, 6))
sns.regplot(data=df, 
            x="Rainfall Impact (Annual Precipitation in mm)", 
            y="Groundwater Depletion Rate (%)", 
            scatter_kws={"alpha": 0.6}, line_kws={"color": "red"})

plt.title("Groundwater Depletion vs. Rainfall Impact")
plt.xlabel("Annual Precipitation (mm)")
plt.ylabel("Groundwater Depletion Rate (%)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

To find out if low rainfall leads to high groundwater depletion? No, This chart suggests that rainfall alone is not a major factor in preventing groundwater depletion. Other factors, such as water consumption, urbanization, and agricultural practices, might be playing a larger role.

```python
# Chart 16:- Countries Most at Risk: High Water Scarcity & High Groundwater Depletion
top_stressed = df.groupby("Country")["Groundwater Depletion Rate (%)"].mean().nlargest(10).index
df_stressed = df[df["Country"].isin(top_stressed)]

fig, ax1 = plt.subplots(figsize=(12,6))

# Bar Chart: Water Scarcity Level
sns.barplot(data=df_stressed, x="Country", y="Groundwater Depletion Rate (%)", ax=ax1, color="red", alpha=0.6)

# Line Chart: Groundwater Depletion Rate
ax2 = ax1.twinx()
sns.lineplot(data=df_stressed, x="Country", y="Groundwater Depletion Rate (%)", ax=ax2, color="black", marker="o")

ax1.set_ylabel("Water Scarcity Level")
ax2.set_ylabel("Groundwater Depletion Rate (%)")
plt.title("Top 10 Most Water-Stressed Countries")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

- Spain, Canada, and Mexico have higher water scarcity levels, suggesting high demand and limited availability.
- South Africa, France, and Russia have lower water scarcity levels, indicating either better water management or lower demand.
- Mexico has the highest depletion rate, exceeding 2.9%.
- Argentina and Spain also have high depletion rates, suggesting unsustainable groundwater use.
- South Africa and France have lower depletion rates, implying better conservation or reliance on alternative water sources.
- Some countries (e.g., Mexico, Argentina, Spain) experience both high water scarcity and high groundwater depletion, indicating a severe water crisis.
- Others (e.g., France, Russia) have low scarcity but still depleting groundwater, which could become problematic in the future.
- The shaded regions suggest significant variability or uncertainty in data, meaning that water scarcity and depletion rates fluctuate over time..
